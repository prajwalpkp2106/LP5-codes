```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import fashion_mnist
```

```python
(x_train,y_train),(x_test,y_test)=fashion_mnist.load_data()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,Conv2D,MaxPooling2D,Flatten
# Conv2D → extracts features from images
# MaxPooling2D → reduces size, keeps important info
# Flatten → converts image data into 1D
# Dense → fully connected layer (final decision making)
# Dropout → prevents overfitting
```

```python
from tensorflow.keras.losses import sparse_categorical_crossentropy
# Defines loss function for multi-class classification
model=Sequential()
```

```python
# Takes a 28×28 grayscale image
# Uses a 3×3 small window to scan the image
# Does this 128 times with different filters
# Each filter finds patterns like:
# edges
# shapes
# Applies ReLU → keeps useful values
```

```python
model.add(Conv2D(128, kernel_size=(3,3), activation='relu', input_shape=(28,28,1)))
model.add(MaxPooling2D())
model.add(Flatten())
# Randomly removes 20% neurons during training
model.add(Dropout(0.2))
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
model.compile(optimizer='adam',metrics=['accuracy'],loss='sparse_categorical_crossentropy')
```

```python
# Trains the CNN on the training data for 5 epochs.
model.fit(x_train,y_train,epochs=5)
```

```python
metrics=model.evaluate(x_test,y_test)
```

```python
metrics
```

```python
#predict label for 5th image
image_index=4
prediction=model.predict(x_test[image_index:image_index+1])
prediction_label=np.argmax(prediction)
true_label=y_test[image_index]
print(f"Predicted label {prediction_label}")
print(f"Actual label: {true_label}")
if(prediction_label==true_label):
    print("Correct Prediction")
else:
    print("Incorrect prediction")
```
